```python
import pandas as pd
import matplotlib.pyplot as plt
import sys


# online retail dataset read in csv file
df = pd.read_csv('online_retail_II.csv')


#Notice in the result that some rows have been removed 
#These rows had cells with empty values.
df.dropna(inplace=True)
print("\n**** print all dataset ****")
print(df.head())

# Returns True for every row that is a duplicate, otherwise False:
df.duplicated()

# To remove duplicates, use the drop_duplicates() method.
df.drop_duplicates(inplace = True)

 # convert to datetime
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

#print the date max - min date
print("\n**** maxmimum date - minimum date ****")
print("\n",df['InvoiceDate'].max()-df['InvoiceDate'].min())

#sum of price 
print("**** sum of price ****")
print("\n",df['Price'].sum())

#load column to relationship between from this dataset 
data = df[['Customer ID','Invoice', 'Description', 'Quantity']]
print("\n**** print the dataset columns : CustomerID, Invoice, Description, Quantity ****")
print("\n",data.head(10))

df[['Quantity','Price']].plot(kind= 'hist')


#save the figure in system
plt.savefig("figure 1.png")
sys.stdout.flush()

print("\n**** Show the diagram on quantity or price ****")
plt.show()
```
